#Yeni eklediÄŸim Ã¶zellikler
Burada hocanÄ±n kullandÄ±ÄŸÄ± yÃ¶ntemlerden ayrÄ± yÃ¶ntemler deniyorum.

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

```

```python
scaler = MinMaxScaler()
df_abnyc2019[['price', 'minimum_nights', 'availability_365']] = scaler.fit_transform(df_abnyc2019[['price', 'minimum_nights', 'availability_365']])

# AÄŸÄ±rlÄ±klandÄ±rma faktÃ¶rleri:
A = 0.6
B = 0.2
C = 0.2

# Puanlama formÃ¼lÃ¼:
df_abnyc2019['score'] = (A * df_abnyc2019['price']) + (B * df_abnyc2019['minimum_nights']) + (C * df_abnyc2019['availability_365'])

# PuanlarÄ± yÃ¼ksekten dÃ¼ÅŸÃ¼ÄŸe doÄŸru sÄ±ralayalÄ±m:
df_abnyc2019 = df_abnyc2019.sort_values(by='score', ascending=False)

print(df_abnyc2019)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df_abnyc2019['neighbourhood_group'], df_abnyc2019['score'], color='blue')
plt.xlabel('neighbourhood_group')
plt.ylabel('Score')
plt.title('En Uygun Konaklama Yerleri')
plt.show()
```

```python
sns.barplot(x="neighbourhood_group",
            y="reviews_per_month",
            hue="number_of_reviews",
            data=df_abnyc2019)
plt.title("Neighbourhood_group - reviews_per_month - number_of_reviews")
```
